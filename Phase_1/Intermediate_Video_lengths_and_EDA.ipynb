# Visualization functions - FOR DEBUGGING / DEMO Purposes:

```python
# for visualization purposes only:

#------------------------------------------------------------------------------------
def video_to_gif(video_path, save_as_gif_path):
    
    # load video
    clip = moviepyEditor.VideoFileClip(video_path)
    
    # get segment from video:
    clip = clip.subclip(0, 5)
    
    # save the video as GIF
    clip.write_gif(save_as_gif_path)
    
video_to_gif(video_path="sample_processed_video.mp4", save_as_gif_path="sample_processed_gif.gif")
```

# Utilities functions:

```python
#------------------------------------------------------------------------------------
def extract_audio_from_video(video_path, save_mp3_audio_as):
    
    # define clip from path:
    clip = moviepyEditor.VideoFileClip(video_path)
    
    # extract audio:
    clip.audio.write_audiofile(save_mp3_audio_as)
    
# Extract audio:
extract_audio_from_video(video_path="../data/Video_chunks/Video_chunks/trial_truth_046_008.mp4", save_mp3_audio_as="sample_audio.mp3")
```

# Find video lengths and save them to csv:

```python
#------------------------------------------------------------------------------------
def get_video_length(video_path):
    # loading video dsa gfg intro video
    clip = moviepyEditor.VideoFileClip(video_path)

    # getting duration of the video
    duration = clip.duration

    # printing duration
    print(f"Duration ({os.path.basename(video_path)}): " + str(duration))
    
    return duration

    # showing final clip
    # clip.ipython_display()
    

def convert_dictionary_to_csv(data_dictionary, save_csv_as):
    # convert data dictionary to dataframe format:
    df = [pd.DataFrame({ k : v }) for k, v in data_dictionary.items()]
    df = pd.concat(df, axis=1)
    # save it:
    df.to_csv(save_csv_as, index=False) 

    
def get_video_lengths(video_path_dir="../data/Video_chunks/Video_chunks", save_csv_as="DARE_video_lengths.csv"):
    # save data to csv:
    dictionary_for_csv = {"video_name": [], "video_path": [], "duration_in_seconds": []}
    
    # go over each video and print their length and save to csv:
    for video_name in os.listdir(video_path_dir):
        # get video data (path and video length in seconds):
        video_key = video_name
        video_path = os.path.join(video_path_dir, video_name)
        video_length = get_video_length(video_path)
        
        # store to dictionary of arrays - to later save into CSV for EDA
        dictionary_for_csv["video_name"].append(video_key)
        dictionary_for_csv["video_path"].append(video_path)
        dictionary_for_csv["duration_in_seconds"].append(video_length)
    
    # save to csv:
    convert_dictionary_to_csv(data_dictionary=dictionary_for_csv,
                              save_csv_as=save_csv_as)
    
    
# get video lengths:
get_video_lengths(video_path_dir="../data/Video_chunks/Video_chunks", save_csv_as="DARE_video_lengths.csv")
```

# Plot video duration histogram:

```python
def plot_duration_histogram(read_csv_path="DARE_video_lengths.csv", plot_title="DARE Video Duration"):
    
    df = pd.read_csv(read_csv_path, index_col=False)
    
    # Creating dataset
    durations = list(df["duration_in_seconds"].values)
    
    # intialize dictionary for visualization of video durations:
    histogram_dictionary = {}
    
    # collect unique lengths:
    for d in durations:
        d = round(d, 1)
        if not (d in histogram_dictionary):
            histogram_dictionary[d] = 0
        histogram_dictionary[d]+=1
    
    # collect series and labels:
    duration_series = []
    label_series = []
    for duration_key in histogram_dictionary:
        duration_series.append(histogram_dictionary[duration_key])
        label_series.append(duration_key)
    # duration_series = pd.Series(duration_series)
    # label_series
        
    # dataframe using frequencies and x_labels from the OP
    df = pd.DataFrame({'Frequency': duration_series}, index=label_series)

    # plot
    ax = df.plot(kind='bar', figsize=(12, 8), title='Amount Frequency',
                 xlabel='Amount ($)', ylabel='Frequency', legend=False)

    # annotate
    ax.bar_label(ax.containers[0], label_type='edge')

    # pad the spacing between the number and the edge of the figure
    ax.margins(y=0.1)

    save_as = plot_title.replace(" ", "_")
    
    
plot_duration_histogram(read_csv_path="DARE_video_lengths.csv", plot_title="DARE Video Duration")
```
