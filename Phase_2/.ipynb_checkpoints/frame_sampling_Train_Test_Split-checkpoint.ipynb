# Collect Video Length from DARE Videos:

```python
import os
import moviepy
import moviepy.editor as moviepyEditor
import pandas as pd

#------------------------------------------------------------------------------------
# PREPARE DATA:
def get_video_info(video_path):
    # loading video dsa gfg intro video
    clip = moviepyEditor.VideoFileClip(video_path)

    # getting duration of the video
    duration = clip.duration
    fps = clip.fps
    total_frames = duration*fps

    # printing duration
    print(f"Duration ({os.path.basename(video_path)}): " + str(duration))
    
    return duration, fps, total_frames

    # showing final clip
    # clip.ipython_display()
    

def convert_dictionary_to_csv(data_dictionary, save_csv_as):
    # convert data dictionary to dataframe format:
    df = [pd.DataFrame({ k : v }) for k, v in data_dictionary.items()]
    df = pd.concat(df, axis=1)
    # save it:
    df.to_csv(save_csv_as, index=False) 

    
def get_video_lengths(video_path_dir="../data/Video_chunks/Video_chunks", save_csv_as="DARE_video_lengths.csv"):
    # save data to csv:
    dictionary_for_csv = {"video_name": [], "video_path": [], "duration_in_seconds": [],
                          "FPS": [], "total_frame_count": []}
    
    # go over each video and print their length and save to csv:
    for video_name in os.listdir(video_path_dir):
        # get video data (path and video length in seconds):
        video_key = video_name
        video_path = os.path.join(video_path_dir, video_name)
        video_length, video_FPS, video_total_frames_num = get_video_info(video_path)
        
        # store to dictionary of arrays - to later save into CSV for EDA
        dictionary_for_csv["video_name"].append(video_key)
        dictionary_for_csv["video_path"].append(video_path)
        dictionary_for_csv["duration_in_seconds"].append(video_length)
        dictionary_for_csv["FPS"].append(video_FPS)
        dictionary_for_csv["total_frame_count"].append(video_total_frames_num)
    
    # save to csv:
    convert_dictionary_to_csv(data_dictionary=dictionary_for_csv,
                              save_csv_as=save_csv_as)
    
    
# get video lengths:
get_video_lengths(video_path_dir="./data/Video_chunks/Video_chunks", save_csv_as="DARE_video_lengths.csv")
```

# Find best distribution for data split:

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import copy


def stratified_sampling_and_plot(df_path,
                                 truth_thresh,
                                 lie_thresh,
                                 save_final_csvs=False):
    
    # input MIN length (in seconds) 
    min_input_length = 3.0
    
    # read in the dataframe with all labels:
    video_lengths_df = pd.read_csv(df_path)
    # find root name of the video, without video count-tag, e.g. trial_truth_"001"_000 becomes trial_truth_"001":
    video_lengths_df['video_root_name'] = video_lengths_df.video_name.str.split('_').str[:3].str.join("_")
    
    # filter out ones that are at least THREE seconds in length:
    video_lengths_df = video_lengths_df.loc[video_lengths_df['duration_in_seconds'] >= min_input_length]
    video_lengths_df.reset_index(drop=True, inplace=True)
    
    total_count = len(video_lengths_df)
    total_frames = sum(video_lengths_df["total_frame_count"].tolist())
    print("- TOTAL SAMPLES:", total_count)
    print("- TOTAL  FRAMES:", total_frames, "\n")
    
    # get only label (truth or lie):
    video_lengths_df['label'] = video_lengths_df.video_name.str.split('_').str[1]
    
    # split them into two different dataframes:
    truth_df = video_lengths_df.loc[video_lengths_df['label'] == "truth"]
    lie_df = video_lengths_df.loc[video_lengths_df['label'] == "lie"]
    
    ##################################
    # Calculate occurences on truth: #
    ##################################
    # get counts per video_root_name for truth and lie:
    truth_root_name_counts = pd.DataFrame(truth_df.video_root_name.value_counts().reset_index().values, columns=["video_root", "sample_count"])
    
    total_truth_samples_num = sum(truth_root_name_counts['sample_count'].tolist())
    truth_thresh = truth_thresh  # 5.0
    # get upper and lower bounds of TRUTH data:
    truth_upper_df = truth_root_name_counts.loc[truth_root_name_counts['sample_count'] >  truth_thresh]
    truth_lower_df = truth_root_name_counts.loc[truth_root_name_counts['sample_count'] <= truth_thresh]
    # get amount of total samples for TRUTH:
    truth_sum_upper = sum(truth_upper_df["sample_count"].tolist())
    truth_sum_lower = sum(truth_lower_df["sample_count"].tolist())
    # get amout of total frames for TRUTH
    truth_frameCounter_upper = 0 # [if (root_name in truth_upper_df["video_root"]) truth_df.iloc[idx]["total_frame_count"] else 0 for idx, root_name in enumerate(truth_df["video_root"].tolist())]
    for idx, root_name in enumerate(truth_df["video_root_name"].tolist()):
        if (root_name in truth_upper_df["video_root"].tolist()):
            truth_frameCounter_upper += truth_df.iloc[idx]["total_frame_count"]
    truth_frameCounter_lower = 0 # [if (root_name in truth_lower_df["video_root"]) truth_df.iloc[idx]["total_frame_count"] else 0 for idx, root_name in enumerate(truth_df["video_root"].tolist())]
    for idx, root_name in enumerate(truth_df["video_root_name"].tolist()):
        if (root_name in truth_lower_df["video_root"].tolist()):
            truth_frameCounter_lower += truth_df.iloc[idx]["total_frame_count"]
    #\/\/\/\/\//\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\/\
    print("--- TRUTH STATS ---")
    truth_frameCounter_upper_PERC = truth_frameCounter_upper/total_frames*100
    print("truth frameCounter % for TRAIN:", truth_frameCounter_upper_PERC)
    truth_frameCounter_lower_PERC = truth_frameCounter_lower/total_frames*100
    print("truth frameCounter % for TEST:", truth_frameCounter_lower_PERC, "\n")
    truth_upper_PERC = truth_sum_upper/total_count*100
    # print(f"truth SAMPLE NUM  % for TRAIN: {truth_sum_upper}/{total_count} = {truth_upper_PERC}")
    truth_lower_PERC = truth_sum_lower/total_count*100
    # print(f"truth SAMPLE NUM  % for TEST: {truth_sum_lower}/{total_count} = {truth_lower_PERC}", "\n")
    # print(f"% of Truth counts >  {truth_thresh}:", ( (truth_sum_upper) / total_truth_samples_num) * 100)
    # print(f"% of Truth counts <= {truth_thresh}:", ( (truth_sum_lower) / total_truth_samples_num) * 100)
    print("-------------------")
    
    ##################################
    # Calculate occurences on lie: #
    ##################################
    lie_root_name_counts = pd.DataFrame(lie_df.video_root_name.value_counts().reset_index().values, columns=["video_root", "sample_count"])
    
    total_lie_samples_num = sum(lie_root_name_counts['sample_count'].tolist())
    lie_thresh = lie_thresh  # 6.0
    # get upper and lower bounds of LIE data:
    lie_upper_df = lie_root_name_counts.loc[lie_root_name_counts['sample_count'] >  lie_thresh]
    lie_lower_df = lie_root_name_counts.loc[lie_root_name_counts['sample_count'] <= lie_thresh]
    # get amount of total samples for LIE:
    lie_sum_upper = sum(lie_upper_df["sample_count"].tolist())
    lie_sum_lower = sum(lie_lower_df["sample_count"].tolist())
    # get amount of total frames for LIE:
    lie_frameCounter_upper = 0 # [if (root_name in lie_upper_df["video_root"]) lie_df.iloc[idx]["total_frame_count"] else 0 for idx, root_name in enumerate(lie_df["video_root"].tolist())]
    for idx, root_name in enumerate(lie_df["video_root_name"].tolist()):
        if (root_name in lie_upper_df["video_root"].tolist()):
            lie_frameCounter_upper += lie_df.iloc[idx]["total_frame_count"]
    lie_frameCounter_lower = 0 # [if (root_name in lie_lower_df["video_root"]) lie_df.iloc[idx]["total_frame_count"] else 0 for idx, root_name in enumerate(lie_df["video_root"].tolist())]
    for idx, root_name in enumerate(lie_df["video_root_name"].tolist()):
        if (root_name in lie_lower_df["video_root"].tolist()):
            lie_frameCounter_lower += lie_df.iloc[idx]["total_frame_count"]
    print("--- LIE STATS ---")
    lie_frameCounter_upper_PERC = lie_frameCounter_upper/total_frames*100
    print("lie frameCounter % TRAIN:", lie_frameCounter_upper_PERC)
    lie_frameCounter_lower_PERC = lie_frameCounter_lower/total_frames*100
    print("lie frameCounter % TEST:", lie_frameCounter_lower_PERC, "\n")
    lie_upper_PERC = lie_sum_upper/total_count*100
    # print(f"lie SAMPLE NUM % for TRAIN: {lie_sum_upper}/{total_count} = {lie_upper_PERC}")
    lie_lower_PERC = lie_sum_lower/total_count*100
    # print(f"lie SAMPLE NUM % for TEST: {lie_sum_lower}/{total_count} = {lie_lower_PERC}", "\n")
    # print(f"% of lie counts >  {lie_thresh}:", ( (lie_sum_upper) / total_lie_samples_num) * 100)
    # print(f"% of lie counts <= {lie_thresh}:", ( (lie_sum_lower) / total_lie_samples_num) * 100)
    print("-----------------")
    
    '''
    truth_total_test_PERC = truth_lower_PERC
    truth_total_test_frameCounter_PERC = truth_frameCounter_lower_PERC
    lie_total_test_PERC = lie_lower_PERC
    lie_total_test_frameCounter_PERC = lie_frameCounter_lower_PERC
    
    
    print("\n")
    print("#####################################################################################")    
    print("--- FINAL DISTRIBUTION STATS ---")
    print(f"TRUTH/LIE TRAIN In Frames: T={truth_frameCounter_upper_PERC} VS. L={lie_frameCounter_upper_PERC} = {(truth_frameCounter_upper_PERC)+(lie_frameCounter_upper_PERC)}")
    print(f"TRUTH/LIE TEST  In Frames: T={truth_total_test_frameCounter_PERC} VS. L={lie_total_test_frameCounter_PERC} = {(truth_total_test_frameCounter_PERC)+(lie_total_test_frameCounter_PERC)}")
    print("\n")
    print(f"TRUTH/LIE TRAIN In %: T={truth_upper_PERC} VS. L={lie_upper_PERC} = {(truth_upper_PERC)+(lie_upper_PERC)}")
    print(f"TRUTH/LIE TEST  In %: T={truth_total_test_PERC} VS. L={lie_total_test_PERC} = {(truth_total_test_PERC)+(lie_total_test_PERC)}")
    print("--------------------------------")
    '''
    
    total_unique_IDs_number = len(video_lengths_df["video_root_name"].unique())
    
    truth_train_IDs = list(set(list(truth_upper_df["video_root"])))
    truth_test_IDs =  list(set(list(truth_lower_df["video_root"])))
    
    lie_train_IDs = list(set(list(lie_upper_df["video_root"])))
    lie_test_IDs =  list(set(list(lie_lower_df["video_root"])))
    
    train_all_IDs = truth_train_IDs + lie_train_IDs
    test_all_IDs  = truth_test_IDs + lie_test_IDs
    
    # get indices for TRAIN selection:
    train_row_idx = []
    test_row_idx = []
    odd_behavior_counter = 0
    
    # go over each row and extract indices that belong to TRAIN and TEST data:
    for curr_idx, row_data in video_lengths_df.iterrows():
        curr_ID = row_data["video_root_name"]
        if curr_ID in train_all_IDs:
            train_row_idx.append(curr_idx)
        elif curr_ID in test_all_IDs:
            test_row_idx.append(curr_idx)
        else:
            odd_behavior_counter+=1
            # print("---")
            # print("Odd behavior:", list(row_data))
            # print("---")
    print("\n@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
        
    train_df = video_lengths_df.loc[video_lengths_df.index[train_row_idx]]
    train_df.reset_index(drop=True, inplace=True)
    
    test_df = video_lengths_df.loc[video_lengths_df.index[test_row_idx]]
    test_df.reset_index(drop=True, inplace=True)
    
    # print(f"TRAIN  FRAME %: {sum(train_df['total_frame_count'].tolist())}/{sum(video_lengths_df['total_frame_count'].tolist())} = {sum(train_df['total_frame_count'].tolist())/sum(video_lengths_df['total_frame_count'].tolist())*100}")
    # print(f"TEST   FRAME %: {sum(test_df['total_frame_count'].tolist())}/{sum(video_lengths_df['total_frame_count'].tolist())} = {sum(test_df['total_frame_count'].tolist())/sum(video_lengths_df['total_frame_count'].tolist())*100}")
    print(f"TRAIN  FRAME %: {sum(train_df['total_frame_count'].tolist())/sum(video_lengths_df['total_frame_count'].tolist())*100}")
    print(f"TEST   FRAME %: {sum(test_df['total_frame_count'].tolist())/sum(video_lengths_df['total_frame_count'].tolist())*100}")
    # print("---")
    # print(f"TRAIN SAMPLE %: {len(train_df)}/{len(video_lengths_df)} = {len(train_df)/len(video_lengths_df)*100}")
    # print(f"TEST  SAMPLE %: {len(test_df)}/{len(video_lengths_df)} = {len(test_df)/len(video_lengths_df)*100}")
    
    print("~> odd_behavior_counter:", odd_behavior_counter)
    print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
    
    
    if save_final_csvs:
        train_df.to_csv('train_DARE_test.csv', index=False)
        test_df.to_csv('test_DARE_test.csv', index=False)
    

for t in [3.0, 4.0, 5.0, 6.0, 7.0]:
    for l in [3.0, 4.0, 5.0, 6.0, 7.0]:
        print("######################")
        print(f"[T={t}] | [L={l}]")
        print("######################")
        stratified_sampling_and_plot(df_path="../CSVs/DARE_video_lengths.csv",
                                     truth_thresh=t,  # 6.0,
                                     lie_thresh=l,  # 6.0)
                                     save_final_csvs=False)
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")


# 6.0 for both=> 69.28 / 30.7 split
# T=5.0, L=6.0 => 73.37 / 26.6 split, bad for testing
# T=6.0, L=4.0 => good distribution for TRAIN/VAL
```

# Generate split based on the above analysis:

```python
def generate_split(t, l):
    print("######################")
    print("Final Save is according to:")
    print(f"[T={t}] | [L={l}]")
    print("######################")
    stratified_sampling_and_plot(df_path="../CSVs/DARE_video_lengths.csv",
                                 truth_thresh=t,  # 6.0,
                                 lie_thresh=l,
                                 save_final_csvs=True)  # 6.0)
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    

'''
######################
[T=6.0] | [L=6.0]
######################
- TOTAL SAMPLES: 743
- TOTAL  FRAMES: 80118.98256583443 

--- TRUTH STATS ---
truth frameCounter % for TRAIN: 32.402848757996225
truth frameCounter % for TEST: 10.928229804557986 

-------------------
--- LIE STATS ---
lie frameCounter % TRAIN: 36.87856255116601
lie frameCounter % TEST: 19.790358886279318 

-----------------

@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
TRAIN  FRAME %: 69.28141130916237
TEST   FRAME %: 30.71858869083726
~> odd_behavior_counter: 0
@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
'''

# Therfore, we will use 6,6 for Truth and Lie:
generate_split(t=6, l=6)
```
