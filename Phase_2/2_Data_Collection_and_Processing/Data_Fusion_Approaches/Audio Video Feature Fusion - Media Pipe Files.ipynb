### Method 1 - Concatenation

This code loads audio and video features from two separate .npy files, normalizes each feature set by subtracting the mean and dividing by the standard deviation, and then concatenates the two feature sets into a single array using NumPy's concatenate() function.
The flatten() method is called on each normalized feature set before concatenating them to create a 1D array, as the concatenate() function requires both input arrays to have the same number of dimensions along all axes except the concatenation axis.
The resulting fused feature is a 1D array that contains both the normalized audio and video features. This can be used for further processing or modeling, such as feeding the feature set into a machine learning algorithm for classification or regression.

```python
import numpy as np

# Load audio and video features from .npy files
audio_feature = np.load("trial_lie_001_000_audio.npy")
video_feature = np.load("trial_lie_001_000_coord.npy")

# Normalize audio and video features
audio_norm = (audio_feature - np.mean(audio_feature)) / np.std(audio_feature)
video_norm = (video_feature - np.mean(video_feature)) / np.std(video_feature)

# Concatenate normalized audio and video features
fused_feature = np.concatenate((audio_norm.flatten(), video_norm.flatten()))
print(fused_feature)

# Save the fused feature as a .csv file
np.savetxt("fused_feature_concatenation.csv", fused_feature, delimiter=",")


# Use the fused feature for further processing or modeling
print("Shape of fused feature:", fused_feature.shape)
```

### Method 2 - PCA 

The code loads the visual features from a numpy array file and standardizes the data using StandardScaler. Then, it performs PCA to reduce the dimensionality of the visual features, keeping only the minimum number of principal components required to explain the variance in the data. The number of principal components is determined by min(visual_features_scaled.shape[0], visual_features_scaled.shape[1]), which ensures that the number of principal components does not exceed the number of samples or features in the input data.

Ensure that the number of rows in the two feature arrays match: Since PCA reduces the dimensionality of the input data, the resulting feature arrays may have different numbers of rows. To concatenate the two feature arrays, the code determines the minimum number of rows between the two arrays using min_rows, and then slices both arrays to keep only the first min_rows rows. This ensures that the two arrays have the same number of rows and can be concatenated along the horizontal axis.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load and preprocess the audio features
audio_features = np.load('trial_lie_001_000_audio.npy')  # assuming numpy array format
audio_scaler = StandardScaler()
audio_features_scaled = audio_scaler.fit_transform(audio_features)
n_components_audio = min(audio_features_scaled.shape[0], audio_features_scaled.shape[1])
audio_pca = PCA(n_components=n_components_audio)  # reduce dimensionality to n_components_audio using PCA
audio_features_pca = audio_pca.fit_transform(audio_features_scaled)

# Load and preprocess the visual features
visual_features = np.load('trial_lie_001_000_coord.npy')  # assuming numpy array format
visual_scaler = StandardScaler()
visual_features_scaled = visual_scaler.fit_transform(visual_features)
n_components_visual = min(visual_features_scaled.shape[0], visual_features_scaled.shape[1])
visual_pca = PCA(n_components=n_components_visual)  # reduce dimensionality to n_components_visual using PCA
visual_features_pca = visual_pca.fit_transform(visual_features_scaled)

# Ensure that the number of rows in the two feature arrays match
min_rows = min(audio_features_pca.shape[0], visual_features_pca.shape[0])
audio_features_pca = audio_features_pca[:min_rows, :]
visual_features_pca = visual_features_pca[:min_rows, :]

# Concatenate the audio and visual features
fused_features = np.concatenate((audio_features_pca, visual_features_pca), axis=1)

np.savetxt("fused_feature_PCA.csv", fused_features, delimiter=",")

print(fused_features)
```

### Method 3 - Early Fusion

In early fusion technique instead of plain stacking of the values and arrays, the data here is combined on raw values and combines it along a particular axis and adjusting the dimesionality of the arrays for eliminating a set of values in order to reshape the arrays to match. Early Fusion of pre-extracted features can capture the correlation between different modalities more explicitly and may be able to leverage this correlation for better performance compared to concatenation.

```python
import numpy as np

# load the audio and video features from the .npy files
audio_features = np.load('trial_lie_001_000_audio.npy')
video_features = np.load('trial_lie_001_000_coord.npy')

# check the shapes of the input arrays
print(audio_features.shape)
print(video_features.shape)

# if the shapes don't match, try to reshape the arrays to match
if audio_features.shape[0] != video_features.shape[0]:
    if audio_features.shape[0] < video_features.shape[0]:
        video_features = video_features[:audio_features.shape[0], :]
    else:
        audio_features = audio_features[:video_features.shape[0], :]

# concatenate the audio and video features horizontally
combined_features = np.hstack((audio_features, video_features))

# perform normalization on the combined features
mean = np.mean(combined_features, axis=0)
std = np.std(combined_features, axis=0)
std += 1e-10 # add a small constant value to avoid division by zero
combined_features = (combined_features - mean) / std

print(combined_features)

np.savetxt("fused_feature_earlyfusion.csv", combined_features, delimiter=",")


# use the combined features for downstream tasks, such as classification
labels = np.load('trial_lie_001_000_label.npy')
#Shape
print("Shape of combined feature:", combined_features.shape)
```

#### Approach 1 

The attention mechanism for both audio and video features is implemented using a fully-connected layer with a 'tanh activation' followed by a 'softmax layer' to compute the attention weights. The attention weights are then used to weight the input features and compute the final attentional features. Finally, the attentional audio and video features are concatenated and passed through a dense layer with softmax activation to perform the classification task. Note that the code assumes that the number of classes is defined as num_classes.

```python
from keras.layers import Input, Dense, LSTM, Concatenate, Permute, Reshape, Lambda, Multiply, Add, Dot
from keras.models import Model
from keras import backend as K

# Load audio and video features from numpy files
audio_features = np.load('trial_lie_001_000_audio.npy')
video_features = np.load('trial_lie_001_000_coord.npy')

# Define input layers for audio and video features
audio_input = Input(shape=(None, audio_features.shape[1]))
video_input = Input(shape=(None, video_features.shape[1]))

# Define attention mechanism for audio features
audio_attention = Dense(1, activation='tanh')(audio_input)
audio_attention = Permute((2, 1))(audio_attention)
audio_attention = Reshape((-1,))(audio_attention)
audio_attention = Lambda(lambda x: x / np.sqrt(audio_features.shape[1]))(audio_attention)
audio_attention = Dense(audio_features.shape[1], activation='softmax')(audio_attention)
audio_attention = Reshape((1, -1))(audio_attention)
audio_attention = Multiply()([audio_input, audio_attention])
audio_attention = Lambda(lambda x: K.sum(x, axis=1))(audio_attention)

# Define attention mechanism for video features
video_attention = Dense(1, activation='tanh')(video_input)
video_attention = Permute((2, 1))(video_attention)
video_attention = Reshape((-1,))(video_attention)
video_attention = Lambda(lambda x: x / np.sqrt(video_features.shape[1]))(video_attention)
video_attention = Dense(video_features.shape[1], activation='softmax')(video_attention)
video_attention = Reshape((1, -1))(video_attention)
video_attention = Multiply()([video_input, video_attention])
video_attention = Lambda(lambda x: K.sum(x, axis=1))(video_attention)

# Concatenate attentional audio and video features
merged_features = Concatenate()([audio_attention, video_attention])
print(merged_features)

# Define output layer for classification or regression
output_layer = Dense(num_classes, activation='softmax')(merged_features)

# Define model with input and output layers
model = Model(inputs=[audio_input, video_input], outputs=output_layer)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model on data
model.fit([audio_features, video_features], labels, epochs=10, batch_size=32)
```
